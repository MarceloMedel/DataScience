### Data Exploration

```python
# ALGORITMO KMEANS
# 1- Se escojen k instancias al azar, que corresponden a los k centroides iniciales
# 2- Se calculan las distancias para el resto de las instancias respecto de los centroides iniciales
# 3- Se evalúan las distancias, reacomodando los centroides si es que existen instancias más cercanas
# 4- Volver al paso 2 y 3 iterativamete hasta que cada instancia quede asignada al cluster del centroide mas cercano


# escogemos el numero optimo de grupos no jerarquizados
# una forma de escoger el numero optimo de clusters es mediante el indice de inercia.
# se define como la suma cuadrática de las distancias entre cada una de las instancias y su centroide más cercano
# por otro lado, el centroide es definido como el centro multidimensional de cada cluster
#definimos un arreglo que almace

# Lista para guardar cada indice de inercia para cada solucion
Inercia = []
#Lista para guardar cada indice de silhouette para cada solucion
Silhouette=[]

#vector con rango de soluciones
K = range(2,20)

#iteracion para cada solucion
for k in K:
    #iniciamos el algoritmo
    km = KMeans(n_clusters=k, init='random', random_state=1312)
    #ajustamos el algoritmo de los datos
    km = km.fit(sample)
    #el metodo fit_predict se ajusta mejor para el calculo del silhouette
    km1 = km.fit_predict(sample) 
    #calculamos silhouette
    sil_prom= silhouette_score(sample, km1)
    #pegamos cada indice de cad solucion en la lista correspondiente
    Inercia.append(km.inertia_)
    Silhouette.append(sil_prom)
    print('Fin iteracion {}'.format(k-1))
```

```python
fig = plt.figure()
sns.set(style="whitegrid",rc={'figure.figsize':(8,5)})
```

```python
#Graficamos inercia  
plt.plot(K, Inercia, 'bx-')
plt.xlabel('k')
plt.ylabel('Suma cuadratica de las distancias')
plt.title('Metodo del codo - elbow - para determinar k óptimo')
plt.show()

#Graficamos Silhouette 
plt.plot(K, Silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('Indice de Silouette')
plt.title('Metodo de Silhouette para determinar k óptimo')
plt.show()
```

```python
# en efecto, la solucion con cinco grupos es la que alcanza la menor inercia y el mayor silhouette
# No obstante, el desempeño general del clustering 
#iniciamos algoritmo para tres grupos
km_final = KMeans(n_clusters=5, init='random', random_state=0)
#ajustamos el algoritmo de los datos
km_final_lab = km_final.fit_predict(sample) 
sample['cluster'] = pd.Series(km_final_lab, index=sample.index)
sample['cluster']=sample['cluster'].add(1) 
sample.head()

#sns.pairplot(data_scal, kind="scatter", hue="cluster", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))

#plt.show()
```

```python
#aplicamos el algoritmo sobre el total de caracteristicas (n_components permite fijar el numero)
pca = PCA().fit(sample)
#Graficamos la varianza explicada acumualada por numero de componentes
plt.figure()
#suma acumulada del ratio de varianza explicada, multiplicado por 100
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
plt.xlabel('Numero de componentes')
plt.ylabel('Varianza explicada (%)') 
plt.title('Analisis PCA caracteristicas comunales')
plt.show()
```
